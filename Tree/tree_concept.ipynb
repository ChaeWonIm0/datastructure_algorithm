#### Tree 구조

> 서로 연결된 Node들이 있는 계층형 구조

> root node와 subtree가 부모<->자식관계로 구성되어 있습니다.

### 개념

* node 노드 : 트리는 다 노드로 구성되어 있어요.
* edge 간선 : 노드간에 연결되어 있는 선
* root node 루트 노드 : 루트 노드가 중요해요. 트리 구조에서 가장 위에있는 노드에요.
* leaf node 리프 노드 : 루트 노드의 반대차원의 노드로, 더이상 아래가 없는 노드에요
* child 자식 노드, parent 부모 노드, sibiling 형제 노드 : 간선으로 연결되어 있는 노드가 있는데, A >> D 로 연결되있을 때, A는 부모노드, D는 자식노드 입니다. B,C와 D는 형제노드 입니다. - 자식, 부모, 형제는 모두 다 고정된 자리가 아니라 상대적인 관계입니다.
* degree 차수 : 각 노드가 갖는 자식의 수입니다. A의 degree는 3, B의 degree는 2입니다. 자식이 없으면 0.
* ancestor 조상노드: 자식노드 입장에서 위쪽으로 간선을 따라가면서 만나는 모든 노드들. F의 조상노드는 B와 A.
* descendant 자손노드 : 한 노드 입장에서 아래쪽으로 간선을 따라가면서 만나는 모든 노드들. D의 자손노드는 G, H와 I.
* height 높이 : root node에서 가장 멀리 떨어져 있는 리프 노드까지의 거리 . 즉, 최대 level 값입니다.
* subtree 서브트리 : 트리의 어떤 노드를 root로 가정한다면, 그 자손으로 구성된 트리를 일컫습니다. B가 루트 노드라면 B : <C, D>로 구성된 서브트리입니다. 

* n진 트리 : 모든 노드의 차수가 n개 이하인 트리. 모든 차수가 3 이하이면 3진 트리로 사용합니다.

```python
class Node:
    def __init__(self, value = 0, left = None, right = None):
        self.value = value
        self.left = left
        self.right = right

class BinaryTree:
    def __init__(self):
        self.root = None

bt = BinaryTree()
bt.root = Node(value = 1)
bt.root.left = Node(value = 2)
bt.root.right = Node(value = 3)
bt.root.left.left = Node(value = 4)
bt.root.left.right = Node(value = 5)
bt.root.right.left = Node(value = 6)
bt.root.right.right = Node(value =7)
```
