### DFS 깊이 우선 탐색 by recursion

#### 모든 노드에 방문해야 합니다.

#### 접근방법
1. preorder 전위순회
2. inorder 중위순회
3. postorder 후위순회

```python
# 이진 트리 세팅
class Node:
    def __init__(self, value = 0, left = None, right = None):
        self.value = value
        self.left = left
        self.right = right
class BinaryTree:
    def __init__(self, value=0):
        self.root = None
        self.value = value
tree = BinaryTree()
tree.root = Node(value='A')
tree.root.left = Node(value = 'B')
tree.root.right = Node(value = 'C')
tree.root.left.left = Node(value = 'D')
tree.root.left.right = Node(value = 'E')
tree.root.right.right = Node(value = 'F')
tree.root.left.left.left = Node(value = 'G')
tree.root.left.left.right = Node(value = 'H')
```

* left child부터 깊이 깊이 접근하게 됩니다.
* 방문은 다른 방식으로 진행


```python
# 아직 선언하지 않음
def dfs(tree_root):
    # root만 전달해도, tree_root가 가리키고 있는 모든 노드에 접근합니다.
    if tree_root is None:
        # 하청의 하청의 하청
        return
    # 각 자식 노드도 서브트리로 인식하고 접근
    # 왼쪽 노드를 확인하라고 하청 ( 재귀 )
    dfs(tree_root.left)
    # 오른쪽 노드를 확인하라고 하청 ( 이또한 재귀 )
    dfs(tree_root.right)
    # 왼쪽 하청, 오른쪽 하청이 모드 끝나면 종료
dfs(tree.root)

```

```python
def dfs(root):
    if root is None:
        return root.value
    dfs(root.left)
    dfs(root.right)
```

#### DFS의 preorder

DFS와 유사한데 , print만 추가해줍니다.

```python
def preorder(node):
    if node is None:
        # 없으면 그 노드 출력
        return
    # 아래 자식노드에 하청을 맡기기 전에 방문해서 인쇄
    print(node.value)
    # 이제부터 하청 시작
    dfs(node.left)
    # 계속 하청의 하청 반복
    dfs(node.right)
```

```python
preorder(tree.root)
# root 노드는 맨앞, 맨 오른쪽 자식 노드는 끝
```

#### inorder 중위 순회

```python
def inorder(node):
    if node is None:
        return
    # 접근은 하지 않고 왼쪽 자식노드에만 계속 하청
    inorder(node.left)
    # 왼쪽 child가 끝나면 방문
    print(node.value)
    # 그 다음 루트에 갔다가 오른쪽 child 방문
    inorder(node.right)
```

```python
inorder(tree.root)
# 그래서 오른쪽 child node는 항상 끝에 출력하게 됩니다
```

#### postorder 후위순회

```python
def postorder(node):
    if node is None:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.value)
```

```python
postorder(tree.root)
# 맨 왼쪽 자식 노드가 맨앞, root 노드가 항상 끝입니다.
```

```python
# DFS 메서드 예제]1
v= 1
def dfs(graph, v, visited):
    visited[v] = True
    print(v, end='')
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i , visited)
```

```python
graph = [[], [2, 3, 8], [1, 7], [1, 4, 5], [3, 5], [3,4], 
         [7], [2 , 6, 8], [1,7]]
visited = [False] * 9
dfs(graph, 1, visited)
```
