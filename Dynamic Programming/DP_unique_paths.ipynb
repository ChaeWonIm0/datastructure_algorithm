### Unique Paths

* 이 로봇은 $ m * n $ grid(격자)위에 있습니다. 로봇은 처음 좌측 상단 모서리 (grid[0][0])에 있고, 우측 하단의 모서리 (grid[m-1][n-1])로 이동하려 합니다. 로봇은 한 번에 오른쪽이나 아래쪽으로만 움직일 수 있습니다.
* 두 정수 m과 n이 주어졌을 때, 로봇이 우측하단 모서리에 도달 가능한 unique paths의 수를 반환하시오.
* 테스트케이스는 답이 $2*10^9$ 이하가 되도록 생성됩니다.

>제약조건
* $ 1<= m, n<= 100$

=  $_6C_3$
= 6 * 5 * 4 /3*2

```python
# deque 없이 어떻게 graph를?
# 내 풀이
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        row = m
        col = n
        path_number = 1
        memo = []
        p = row*col
        memo = [[0]*col for i in range(row)]
        if row + col <= 3:
            return 1
        def DP(n):
            if n < 2:
                return 1 
            for i in range(p):
                memo[i] = memo[i-1]+memo[i-2]-1
            return memo[n]
        
        DP(p)

        return path_number
```

### 7 x 3 격자의 조합 갯수 
#### = 아래 화살표 2개와 오른쪽 화살표 6개로 가능한 조합의 수
##### 총 8개 중에서 2개를 선택해서 특정한 자리에 나열하는 방법  = $_8C_2$

= $8!/(2!*6!)$ $=28$

```python
# DP Top-down 방식

class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        memo = {} # dict 버전
        unique_paths = 0
        def DFS(r,c):
            if r==0 and c == 0:
                memo[(r,c)] = 1
                return memo[(r,c)]
            unique_paths = 0
            if (r,c) not in memo:
                if r-1 >= 0:
                    unique_paths += DFS(r-1, c)
                if c-1 >= 0:
                    unique_paths += DFS(r, c-1)
                memo[(r,c)] = unique_paths
            return memo[(r,c)]
        return DFS(m-1,n-1)
```

```python
s = Solution
s.uniquePaths(s, 3, 7)
```

```python
#  for문 버전
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        memo = [[0]*n for i in range(m)]
        for r in range(m):
            memo[r][0] = 1
        for c in range(n):
            memo[0][c] = 1
        for r in range(1,m):
            for c in range(1, n):
                memo[r][c] = memo[r-1][c] + memo[r][c-1]
        return memo[m-1][n-1]
```
