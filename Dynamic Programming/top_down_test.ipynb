```python
# 문제에서 제공
# 내 풀이
class Solution:
    def climbStairs(self, n: int) -> int:
        memo = {}
        def sum(n):
            if n==1 or n==2:
                return 1

            if n not in memo:
                memo[n] = sum(n-1)+sum(n-2)
            return memo[n]

        return sum(n+1)

s = Solution
s.climbStairs(s, 3)
```

> 하위 문제에 대한 답을 계산한다.

> 하위 문제에 대한 답으로 원래 문제에 대한 답을 낸다

```python
# top-down식
class Solution:
    def climbStairs(self, n: int) -> int:
        memo = {}
        def cs(n):
            if n == 1:
                return 1
            if n == 2:
                return 2
            if n not in memo:
                memo[n] = cs(n-1)+cs(n-2)
            return memo[n]
        return cs(n)
            
```
