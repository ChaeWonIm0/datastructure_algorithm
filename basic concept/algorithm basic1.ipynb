##### 어떤 자료구조를 선택하느냐에 따라, 시간이나 오류해결이 달라지게 됩니다.


한 문제를 다양한 방식으로도 풀어보고, 다른 사람은 어떻게 접근하는지도 찾아봐야 합니다.

##### 알고리즘이란
* 문제해결의 방법 : 어떠한 문제를 풀어내기 위해 정해진 일련의 방법입니다.
> 그중에서 이름을 붙인 것.
* BFS, DFS, binary search, dijkstra 등등

* 하지만 각 문제에 적합한 알고리즘을 선택해야 합니다.
* 알고리즘을 평가할 줄 알아야 합니다.

#### 평가기준

##### 시간복잡도
* 이것만 봅니다

##### 구현복잡도

##### 공간복잡도;

* 시간복잡도와 공간복잡도는 보통 trade-off 관계입니다. 실행시간과 메모리가 반비례관계가 된다는 거죠.


#### 실행시간 (Running time)

##### 프로그램이 시작되는 시점부터, 모든 코드를 처리하는데 걸리는 시간입니다.


##### 반복문의 실행시간
* 1회시 시간 * 반복횟수
* 이중 반복문의 경우 = $(1회시간 * 내부for문의반복횟수)^2+ 1회시간* 반복횟수$

* Big-O표기법

상수 시간복잡도 = $O(1)$로 점근표기
ex) C, T 등 (runtime(n))

최고차항이 1차함수인 시간 복잡도 = $O(n)$으로 점근표기
* ex) an + b, ax + c = $O(n)$

최고차항이 2차함수인 시간 복잡도 = O($n^2$)로 표기

### 시간복잡도의 case

ex) 공 5개를 정렬하는 방법
* 가장 높은 시간복잡도 : worst case
* 가장 낮은 시간복잡도 : best case
* 시간복잡도의 평균값 : average case

* 보통의 알고리즘 시간복잡도는 average case를 말합니다
다만 worst case, average case가 같은 경우가 많아서, 비교적 구하기 쉬운 worst case를 구하면 좋습니다.
