```python
# 해결방식
import heapq

class Solution:
    def networkDelayTime(self, times: List[List[int]], n: int, k: int) -> int:
        
        # Create graph
        adj = [[] for i in range(n + 1)]
        for it in times:
            adj[it[0]].append((it[1], it[2]))

        # inital config
        q = [(0, k)]
        dist = [float("inf") for i in range(n + 1)]
        dist[k] = 0

        # dijkstra algorithm
        while q:
            time, node = heapq.heappop(q)
            for it in adj[node]:
                adjNode, adjTime = it
                if time + adjTime < dist[adjNode]:
                    dist[adjNode] = time + adjTime
                    heapq.heappush(q, (dist[adjNode], adjNode))

        return max(dist[1:]) if max(dist[1:]) != float("inf") else -1
```
